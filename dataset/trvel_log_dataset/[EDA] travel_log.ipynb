````python
TC_CODEA = pd.read_csv(
    'data/travel_log/TL_csv/tc_codea_코드A.csv', encoding='utf-8')

display(TC_CODEA.info())

TC_CODEA
````

- 각 코드의 의미를 보면 활동 구분을 `cd_a`로 하고 있으며, 각 코드의 의미는 `cd_nm`에서 설명하고 있음을 알 수 있음


````python
def print_cd_info(code):
    """
    특정 코드 값에 대한 정보를 출력하는 함수
    """
    global TC_CODEA  # 전역 변수로 선언

    if code.lower() != 'exit':
        # 코드 값에 해당하는 행 선택
        selected_row = TC_CODEA[TC_CODEA['cd_a'] == code]

        # 해당 코드 값에 대한 정보 출력
        if not selected_row.empty:
            code_name = selected_row['cd_nm'].iloc[0]  # cd_nm 컬럼 값
            memo = selected_row['cd_memo'].iloc[0]  # cd_memo 행 값

            print(f"코드: {code}")
            print(f"코드명: {code_name}")
            print(f"메모: {memo}")
        else:
            print(f"코드 '{code}'에 대한 정보가 없습니다.")
````

````python
code = 'STA'
print_cd_info(code)
````

- 전체 여행로그 DB에 사용되는 코드명세에 사용되는 코드들을 정리해둔 파일

````python
TC_CODEB = pd.read_csv(
    'data/travel_log/TL_csv/tc_codeb_코드B.csv', encoding='utf-8')


display(TC_CODEB.info())

TC_CODEB.head()
````

- `code_a` 에서 정의한 각 코드에 연결된 코드들을 정리해둔 일종의 테이블 명세서
- `cd_a`를 통해 `code_a` 의 어떤 코드 하위에 있는 코드인지 알 수 있고, `cd_b` 컬럼을 통해 각 `cd_nm`의 고유 코드를 알 수 있다.
- 두 데이터프레임을 합쳐서 `cd_a`를 기준으로 어떤 값을 나타내기 위한 코드인지 확인하기 쉽게 가공


### CODE_A + CODE_B


````python
# 코드_a의 cd_nm을 'code_nm_a'로 이름을 변경하여 복사
code_a_copy = TC_CODEA.rename(
    columns={'cd_nm': 'code_nm_a', 'cd_memo': 'cd_memo_a'}).copy()

# 코드_b의 cd_nm을 'code_nm_b'로 이름을 변경하여 복사
code_b_copy = TC_CODEB.rename(
    columns={'cd_nm': 'code_nm_b', 'cd_memo': 'cd_memo_b'}).copy()

# 코드_a와 코드_b 데이터프레임을 cd_a를 기준으로 병합
merged_df = pd.merge(code_b_copy, code_a_copy, on='cd_a', how='left')

# 코드 값에 따라 cd_nm_a, cd_nm_b, cd_a, cd_b를 그룹바이하여 출력하는 작업
basic_code = merged_df.groupby('cd_a').apply(
    lambda x: x[['cd_b', 'code_nm_a', 'code_nm_b']].reset_index(drop=True))
# 결과 출력
basic_code
````

- code_a와 code_b를 병합하고 `code_a`를 기준으로 필요컬럼만 group by 한 결과 각 코드의 쓰임새를 더 명확하게 파악할 수 있음.
- 가령 `VIS13` 이되면 여행객에 체험활동 관광지를 방문했다는 것이 됨.
- 하위에 기록할 다양한 데이터들의 상세 항목(카테고리인 경우) 위 데이터셋을 통해 대략적인 내용 파악이 가능함.

### 학습용 feature 구축 방안

- 모델 테스트에 사용된 최종 데이터셋을 만들기 위한 예제 코드
- 다만 `TRAVEL_MISSION_PRIORITY` 의 경우 어떤 로직으로 계산을 했는지 제공된 문서상으론 파악이 어렵습니다.
    - `TN_TRAVEL` 에서 `TRAVEL_MISSION` 을 분석해 특정 로직으로 우선순위를 정해야 합니다.
    - 가장 간단한 방법으로는 각 여행자별로 `1;21;22;24;27;` 같은 방식으로 여행 선호 정보가 입력되어 있습니다.
    - 이는 사용자가 과거에 했던 여행을 선택했던 이유를 표현한 것으로 보여집니다.
    - 따라서, 이 데이터를 분석해 사용자의 성향을 파악해 `TRAVEL_MISSION_PRIORITY` 로 입력하는 것이 가장 합리적이라 생각 됩니다.
    - 이러한 점을 고려할 때  아래의 2가지 방법이 가장 합리적이며 data-driven한 방법이라 생각 됩니다.

- 1. 데이터 분석을 통한 선호도 파악
    - 사용자가 과거에 선택한 여행 목적의 데이터를 분석하여, 가장 자주 선택되거나 평점이 높은 여행 목적을 우선순위로 설정하는 방법. 
    - 이는 사용자의 실제 행동 데이터를 반영하므로, 개인화된 추천에 유리합니다.

- 2. 여행 목적별 평균 평점 분석
    - 사용자 리뷰나 평점을 기반으로 각 여행 목적의 평균 평점을 계산하고, 이를 우선순위 결정의 기준으로 활용하는 방법. 
    - 이는 높은 평점을 받은 여행 목적을 우선적으로 추천하게 되어 여행자의 경험이 강하게 반영 됩니다.

- `TRAVEL_MISSION_PRIORITY`  를 제외한 다른 feature들의 경우 아래의 코드와 같이 생성이 가능합니다.

```python

import pandas as pd

# 원본 df에서 바로 찾을 수 있는 features
selected_columns = TN_TRAVELLER_MASTER[['GENDER', 'AGE_GRP', 'INCOME', 'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8', 'TRAVEL_MOTIVE_1', 'TRAVEL_NUM', 'TRAVEL_COMPANIONS_NUM']]

# 별도의 계산이 필요한 features
# RESIDENCE_TIME_MIN_mean, RCMDTN_INTENTION_MEAN, REVISIT_YN_mean, REVISIT_INTENTION_MEAN 컬럼을 계산하기 위한 코드
# TN_VISIT_AREA_INFO 데이터셋에서 필요한 계산을 수행
TN_VISIT_AREA_INFO['REVISIT_YN_numeric'] = TN_VISIT_AREA_INFO['REVISIT_YN'].apply(lambda x: 1 if x == 'Y' else 0)

# 각 여행 ID별로 필요한 평균값을 계산합니다.
visit_area_aggregations = {
    'RESIDENCE_TIME_MIN': 'mean',
    'RCMDTN_INTENTION': 'mean',
    'REVISIT_YN_numeric': 'mean',
    'REVISIT_INTENTION': 'mean'
}
visit_area_means = TN_VISIT_AREA_INFO.groupby('TRAVEL_ID').agg(visit_area_aggregations).reset_index()
visit_area_means.rename(columns={'RESIDENCE_TIME_MIN': 'RESIDENCE_TIME_MIN_mean', 'RCMDTN_INTENTION': 'RCMDTN_INTENTION_MEAN', 'REVISIT_YN_numeric': 'REVISIT_YN_mean', 'REVISIT_INTENTION': 'REVISIT_INTENTION_MEAN'}, inplace=True)

# TRAVEL_COMPANIONS_NUM_mean 컬럼을 계산하기 위한 코드
# TN_COMPANION_INFO 데이터셋에서 각 여행 ID별로 동반자 수의 평균을 계산

companion_counts = TN_COMPANION_INFO.groupby('TRAVEL_ID').size().reset_index(name='TRAVEL_COMPANIONS_NUM_count')
companion_mean = companion_counts['TRAVEL_COMPANIONS_NUM_count'].mean()
TN_TRAVELLER_MASTER['TRAVEL_COMPANIONS_NUM_mean'] = companion_mean

merged_df = pd.merge(TN_TRAVEL, TN_VISIT_AREA_INFO, on='TRAVEL_ID', how='inner')
```

    - `코드A`, `코드B`의 경우 실제 여행관련 데이터에 사용된 값들이 어떤 의미를 지니는지를 추론하기 위해 사용됩니다.
    - `여행객 Master`, `시군구코드`, `여행`, `동반자정보`, `방문지 정보` 는 실제 추천 시스템을 만들기 위한 더미 데이터로 활용 가능합니다.
    - `관광사진` 의 경우는 `유사 이미지 검색` 기능의 테스트를 위해 사용이 가능합니다.
